**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

```python
#creating a groupby item of the average price by borough and year
dfg = properties_melted.groupby(['Borough', 'Year'])['Average_price'].mean()
dfg
```

```python
#resetting index
dfg = pd.DataFrame(dfg).reset_index()
dfg
```

```python
#preparing to filter the dataframe by removing all rows for non-boroughs

price_ratios = {}
dfg['Borough'].unique()
len(dfg['Borough'].unique())
boroughs = ['Westminster', 'Kensington & Chelsea', 'Hammersmith & Fulham', 'Wandsworth', 'Lambeth', 'Southwark', 'Tower Hamlets', 'Hackney', 'Islington',
           'Camden', 'Brent', 'Ealing', 'Hounslow', 'Richmond upon Thames', 'Kingston upon Thames', 'Merton', 'Sutton', 'Croydon', 'Bromley', 'Lewisham',
           'Greenwich', 'Bexley', 'Havering', 'Barking & Dagenham', 'Redbridge', 'Newham', 'Waltham Forest', 'Haringey', 'Enfield', 'Barnet', 'Harrow',
           'Hillingdon']
boroughs
type(boroughs)
```

```python
#filtering the dataframe to only have rows for boroughs

pd.options.display.float_format = '{:.2f}'.format
dfg = dfg[dfg.Borough.isin(boroughs)]
dfg
```

```python
#checking out the dataframe for a particular borough

b = dfg[dfg.Borough == 'Enfield']
b
b[b['Year'] == 1998]['Average_price']
type(b[b['Year'] == 1998]['Average_price'])
float(b[b['Year'] == 1998]['Average_price'].iloc[0])

#creating the price ratio function

def create_price_ratio(b):
    price_1998 = float(b[b['Year'] == 1998]['Average_price'].iloc[0])
    price_2018 = float(b[b['Year'] == 2018]['Average_price'].iloc[0])
    return price_2018/price_1998
```

```python
#creating a dictionary which contains the price ratio for every borough using a for loop

for i in boroughs:
    b = dfg[dfg['Borough'] == i]
    ratio = create_price_ratio(b)
    price_ratios[i] = ratio 

price_ratios
price_ratios.keys()
price_ratios.values()
```

```python
#converting the price ratio dictionary into a dataframe

price_ratios.keys()
list(price_ratios.keys())

data = {'Borough':list(price_ratios.keys()), 'Price Ratio':list(price_ratios.values())}
df = pd.DataFrame(data)
df
```

```python
# sorting the dataframe of price ratio in descending order and selecting the observations with the biggest ratios

df = df.sort_values(by = ['Price Ratio'], ascending = False)
df = df.reset_index(drop = True)
df
df_top10 = df.head(10)
df_top10
```

```python
#visualizing the price increases for the boroughs that have seen the biggest percentage increase

ax = df_top10[['Borough', 'Price Ratio']].plot(kind = 'bar')
ax.set_xticklabels(df_top10['Borough']);
```
